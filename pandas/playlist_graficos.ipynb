# **Gráficos**

## **Utilizando apenas a biblioteca do pandas**

Criar gráficos diretamente através da biblioteca **Pandas** no Python é uma forma eficiente de visualizar dados sem a necessidade imediata de importar outras bibliotecas complexas, pois o Pandas já integra funcionalidades de visualização.

### Entendendo os Eixos X e Y
Os eixos **X (horizontal)** e **Y (vertical)** são as coordenadas fundamentais de um gráfico que representam as variáveis dos seus dados,. No Pandas, a função `.plot()` identifica automaticamente as colunas do seu DataFrame para preencher esses eixos. Quando você deseja especificar colunas diferentes do índice padrão, deve usar os parâmetros **`x=`** e **`y=`**,.

Para **editar os eixos**, como alterar os valores exibidos ou formatar os rótulos, você pode utilizar funções da biblioteca Matplotlib (que serve de base para o Pandas), como o `plt.xticks()`, que permite configurar a posição e o texto dos labels.

### Tipos de Gráficos (`kind=`)
O parâmetro **`kind`** define o formato visual do gráfico. Os tipos mais comuns incluem,,,:
*   **`bar`**: Gráfico de barras verticais,.
*   **`barh`**: Gráfico de barras horizontais.
*   **`hist`**: Histograma para ver a distribuição de frequências,.
*   **`scatter`**: Gráfico de dispersão para comparar duas variáveis numéricas.
*   **`box`**: Gráfico de caixa para visualizar a distribuição e outliers.
*   **`kde`**: Gráfico de densidade.

### Múltiplos Gráficos com `subplots` e `layout`
Se você possui várias categorias e deseja visualizá-las separadamente, pode usar o parâmetro **`subplots=True`**. Para organizar como esses gráficos aparecem na tela, utiliza-se o **`layout`**, que recebe uma tupla indicando o número de linhas e colunas (por exemplo, `layout=(3, 3)` para organizar nove gráficos em uma grade de 3x3).

### Ajustando o Tamanho (`figsize`)
Para garantir que o gráfico não fique muito pequeno ou esquisito, utiliza-se o parâmetro **`figsize`**. Ele recebe uma tupla com os valores de **comprimento e altura** da figura (ex: `figsize=(15, 10)`), permitindo criar visualizações mais largas ou mais altas conforme a necessidade,.

### Compartilhamento de Eixos (`sharex` e `sharey`)
Ao trabalhar com subplots, é comum que cada gráfico tenha uma escala própria, o que dificulta a comparação direta. Os parâmetros **`sharex=True`** e **`sharey=True`** forçam todos os gráficos a compartilharem a mesma escala nos eixos X e Y, permitindo identificar com clareza quais categorias possuem maior volume ou variação.

### Barras Empilhadas e Organização
Para visualizar a composição de um total entre diferentes categorias em um gráfico de barras, você pode usar o parâmetro **`stacked=True`**. Isso cria **barras empilhadas**, onde as categorias são colocadas umas sobre as outras em vez de lado a lado. Além disso, você pode organizar seus dados antes da plotagem usando funções como `sort_values()` para ordenar as barras de forma crescente ou decrescente.

### Rotacionamento de Eixo
Muitas vezes, os nomes no eixo X são longos e acabam ficando inclinados ou sobrepostos. Para corrigir isso, o Pandas oferece o parâmetro **`rot`** dentro da função de plotagem (ex: `rot=0`), que define o grau de rotação dos rótulos para facilitar a leitura.

## **MATPLOTLIB**

O **Matplotlib** é a biblioteca fundamental para a criação de visualizações de dados em Python, servindo como a base técnica sobre a qual outras ferramentas populares, como **Seaborn e Pandas**, são construídas. Ele permite gerar desde gráficos simples até composições complexas e altamente personalizadas, oferecendo controle total sobre elementos como eixos, legendas e o layout de múltiplos subplots através de seu módulo `pyplot`.

## **Como salvar seu gráfico**

Para salvar os gráficos gerados em Python, existem diferentes métodos que variam dependendo da biblioteca utilizada e do formato de saída desejado, seja uma imagem estática ou um arquivo interativo.

Para gráficos criados com o **Matplotlib**, o processo é muito prático: utiliza-se a função **`plt.savefig()`**, passando entre parênteses o caminho da pasta e o nome do arquivo com a extensão (por exemplo, `teste.png`). Um ponto de extrema importância é que esse comando deve ser executado na **mesma célula** de código em que o gráfico foi chamado.

Caso você esteja utilizando o **Seaborn**, é necessário primeiro atribuir o gráfico a uma variável (por exemplo, `meu_plot = sns.barplot(...)`). A partir dessa variável, você deve chamar a função **`meu_plot.get_figure()`** para capturar a figura e, em seguida, aplicar o método **`.savefig()`**. Se desejar que a imagem salva tenha uma dimensão maior, basta aumentar o tamanho do gráfico antes de salvá-lo.

Se o objetivo for preservar a **interatividade** (como funções de zoom e exibição de dados ao passar o mouse), o ideal é salvar o gráfico em formato **HTML**. Bibliotecas como o **Plotly** oferecem a função **`fig.write_html()`** para essa finalidade. Além disso, você pode exportar o **notebook inteiro do Jupyter para HTML** através do menu "Download as", ou utilizar o comando de terminal `jupyter nbconvert` com o parâmetro `--no-input` para gerar um relatório limpo que **oculta os códigos** e exibe apenas os resultados e gráficos interativos.

### **Configuração do eixo X**

Para configurar o eixo X utilizando a biblioteca **Matplotlib**, a principal função utilizada é a **`plt.xticks()`**.

Abaixo estão as principais formas de ajuste mencionadas nas fontes:

*   **Rotação dos Rótulos:** Se os nomes no eixo X estiverem amontoados ou inclinados, você pode usar o parâmetro **`rotation`** dentro de `plt.xticks()`. Definir `rotation=0` deixa os textos na horizontal, mas você pode usar outros ângulos conforme a necessidade.
*   **Ajuste de Posições e Nomes:** A função `plt.xticks()` aceita dois parâmetros fundamentais: os **ticks** (posições numéricas no eixo, como `0, 1, 2...`) e os **labels** (os nomes que serão escritos nessas posições). Para gráficos de categorias, as posições podem ser definidas pelo comprimento da lista de nomes usando algo como `range(len(labels))`.
*   **Quebra de Linha em Nomes Longos:** Uma técnica para evitar que textos longos invadam o espaço alheio é utilizar a biblioteca **`textwrap`**. Com ela, você pode formatar os rótulos para que quebrem a linha (adicionando um `\n`) após um número específico de caracteres.
*   **Compartilhamento de Eixo (Subplots):** Ao criar múltiplos gráficos simultâneos, você pode usar o parâmetro **`sharex=True`** para que todos os subplots utilizem a mesma escala e alinhamento no eixo X, facilitando a comparação. Caso queira que cada gráfico tenha seu próprio eixo independente, utiliza-se `sharex=False`.
*   **Limpeza da Exibição:** Para evitar que o Python exiba códigos de descrição do objeto acima do gráfico, utiliza-se o comando **`plt.show()`** ao final da configuração.

Além dessas configurações manuais, ao utilizar o **Pandas** para plotar, você pode passar o parâmetro **`rot`** diretamente na função `.plot()` para controlar a rotação do eixo X de forma simplificada.

## **Mover as legendas**

Utilizando a função **`plt.legend()`** e seus parâmetros de posicionamento.

As principais formas de ajustar a posição são:

### 1. Utilizando o parâmetro `loc`
O parâmetro `loc` define em qual extremidade do gráfico a legenda será fixada. Você pode usar strings como:
*   **'upper right'** (superior direito)
*   **'upper left'** (superior esquerdo)
*   **'lower right'** (inferior direito)
*   **'lower left'** (inferior esquerdo)
*   **'center'** (centro).

Existem também números que representam cada uma dessas posições na documentação da biblioteca.

### 2. Ajuste fino com `bbox_to_anchor`
Se você deseja mover a legenda para fora da área do gráfico ou para uma posição muito específica, utiliza-se o **`bbox_to_anchor`**. Ele recebe uma tupla com dois parâmetros **(x, y)** que deslocam a legenda a partir do ponto definido no `loc`:
*   **Eixo X:** Valores maiores que 1 movem a legenda para a direita do gráfico (ex: `1.05`).
*   **Eixo Y:** Ajusta a altura da legenda.

**Exemplo prático:** Para colocar a legenda totalmente fora do gráfico, à direita, você pode usar:
`plt.legend(loc='upper left', bbox_to_anchor=(1, 1))`.

### 3. Organização em colunas (`ncol`)
Se a sua legenda for muito longa e estiver atrapalhando a visualização, você pode transformá-la em múltiplas colunas usando o parâmetro **`ncol`**. Por exemplo, `ncol=3` distribuirá os itens da legenda em três colunas horizontais, o que é útil para posicioná-la no topo ou no rodapé do gráfico.

Essas configurações são universais para as bibliotecas de visualização em Python, pois o **Seaborn** e o **Pandas** rodam sobre o motor do **Matplotlib**. Caso esteja usando o `FacetGrid` do Seaborn, você também pode usar a função específica `g.add_legend()` para adicionar a legenda automaticamente do lado de fora.

## **Alrerando as cores das barras**

Para alterar a cor das barras em seus gráficos utilizando Python, existem diversas abordagens que variam desde mudanças simples até a aplicação de lógica condicional para destacar dados específicos.

### 1. Alteração Simples e Uniforme
A maneira mais básica de mudar a cor de todas as barras (como no `sns.barplot`) é através do parâmetro **`color`**. Você pode definir uma cor única para todo o gráfico utilizando nomes de cores (ex: 'blue') ou abreviações (ex: 'b' para azul).

### 2. Cores Baseadas em Categorias
Para diferenciar as barras conforme uma variável categórica (como "Sexo" ou "Curso"), utiliza-se o parâmetro **`hue`** no Seaborn. Essa função atribui automaticamente cores distintas para cada grupo e gera uma legenda correspondente. Se estiver utilizando o Pandas com o motor de plotagem interativo do Plotly, o parâmetro equivalente para essa função é o **`color`**.

### 3. Uso de Paletas e Listas Personalizadas
O parâmetro **`palette`** é utilizado para aplicar conjuntos de cores. Você pode:
*   Usar **paletas pré-configuradas** da biblioteca, como 'Blues'.
*   Passar uma **lista customizada** de cores (ex: `['red', 'black']`) ou **códigos hexadecimais**. Se a lista for menor que o número de barras, as cores serão repetidas sequencialmente.

### 4. Destaque Condicional (Cores por Valor)
Para destacar barras que atingiram uma meta, você pode usar uma **lista de compreensão (list comprehension)** para criar uma paleta dinâmica. Por exemplo, você pode definir que a cor será 'azul' se o valor for maior que 10 e 'vermelho' caso contrário, passando essa lista resultante para o parâmetro `palette`. Para lógicas mais sofisticadas, o ideal é criar uma **função** que receba o valor e retorne a cor, o que torna o código mais organizado e legível.

### 5. Ajustes Adicionais
*   **Ordem das Cores:** No Seaborn, você pode alterar a ordem das cores e da legenda usando o parâmetro **`hue_order`**.
*   **Gráficos da Diagonal:** Em gráficos de matriz (como o `pairplot`), é possível alterar especificamente a cor e o tipo dos gráficos da diagonal (como mudar de densidade para histograma) usando o `diag_kind`.

Ao utilizar essas técnicas, é possível transformar um gráfico comum em uma ferramenta de análise visual muito mais eficiente, **ordenando as barras e destacando pontos críticos** de forma clara.
