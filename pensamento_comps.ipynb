```python
# BIBLIOTECA
import httpx
from bs4 import BeautifulSoup
```

```python
from re import findall

paginas = ["https://averdade.org.br/category/brasil/",
           "https://averdade.org.br/category/brasil/page/2/",
           ]

def acessar_pagina (link):
  #print (f"Estou na p√°gina: {link}")
  pag_web = httpx.get(link)
  bs = BeautifulSoup(pag_web, "html.parser")
  return bs


#loop
for pagina in paginas:
  pagina_inteira = acessar_pagina(pagina)
  # Primeiro, encontra a div principal onde as not√≠cias est√£o contidas
  main_content_div = pagina_inteira.find("div", attrs={"id" : "tdi_50"})

  lista_noticias = [] # Inicializa a lista para cada p√°gina
  if main_content_div:
      h3_elements = main_content_div.find_all("h3", class_="entry-title td-module-title")
      for h3 in h3_elements:
          link_tag = h3.find("a")
          if link_tag and "href" in link_tag.attrs:
              lista_noticias.append(link_tag["href"])

  print(lista_noticias)
```
