## **.loc & .iloc**

###.**loc**  
```
df.loc[:,’col1’:’col3’]
```
Este código filtra as colunas que deseja-se exibir, aqui está o que cada parte faz:

*   **`df.loc`**: É um método do Pandas que permite selecionar dados baseando-se em **rótulos** (nomes de linhas ou colunas).
*   **`:` (antes da vírgula)**: Este caractere isolado indica que você quer selecionar **todas as linhas** do DataFrame.
*   **`,` (vírgula)**: Separa a seleção de linhas (à esquerda) da seleção de colunas (à direita).
*   **`'col1':'col3'`**: Define um **intervalo de colunas** que será retornado. O Pandas selecionará a 'col1', a 'col3' e todas as colunas que estiverem localizadas entre elas na ordem atual do DataFrame.

###**iloc**
O **`iloc`** é um método do Pandas utilizado para selecionar dados baseando-se em sua **posição numérica** (índice inteiro), diferentemente do `loc`, que utiliza os nomes das colunas ou rótulos.

O exemplo a seguir inverte a ordem de exibição das colunas. Aqui está a explicação detalhada do código **`df.iloc[:, ::-1]`**:

*   **`df.iloc`**: Indica que a seleção será feita por índices numéricos (0, 1, 2...).
*   **`:` (antes da vírgula)**: Significa que você está selecionando **todas as linhas** do DataFrame.
*   **`,` (vírgula)**: Separa a regra de seleção das linhas da regra de seleção das colunas.
*   **`::-1` (depois da vírgula)**: Esta é uma técnica de fatiamento (*slicing*) do Python. O sinal de negativo (`-1`) indica que a contagem deve ser feita de trás para frente. No contexto do Pandas, isso é usado para **alterar a ordem das colunas**, invertendo-as completamente.

# **Append Concat Merge & Join**

*   **Concat**: É uma função versátil utilizada para **unir tanto linhas quanto colunas** de diferentes DataFrames. Ela "empilha" os dados ao longo de um eixo definido.
*   **Append**: É um método focado especificamente na **união de linhas**. Ele é útil quando você deseja adicionar novos registros (linhas) ao final de um DataFrame já existente.
*   **Merge**: Considerada a **melhor e mais flexível forma de unir DataFrames**, o `merge` permite combinar dados baseando-se em colunas comuns (chaves), funcionando de forma muito semelhante ao "JOIN" da linguagem SQL.
*   **Join**: É outra ferramenta para **unir DataFrames**, sendo frequentemente utilizada para combinar dados que compartilham o mesmo índice.

A escolha entre qual usar depende se você deseja apenas anexar dados (como no `append` e `concat`) ou se precisa realizar cruzamentos de informações baseados em identificadores comuns (como no `merge` e `join`).

## Concat

O **`pd.concat`** pode "empilhar" DataFrames tanto verticalmente (um abaixo do outro) quanto horizontalmente (lado a lado).

### 1. `ignore_index`
Por padrão, o `concat` preserva os índices originais dos DataFrames que estão sendo unidos.
*   **`ignore_index=False` (Padrão):** O DataFrame resultante manterá os índices de origem (o que pode gerar índices duplicados, como dois registros com índice "0").
*   **`ignore_index=True`:** O Pandas ignora os índices originais e cria um novo índice sequencial (0, 1, 2...) para o DataFrame resultante. É muito útil quando os índices antigos não têm um significado especial e você quer apenas uma contagem limpa.

### 2. `join` ('inner' vs 'outer')
Este parâmetro define como o Pandas lida com colunas (ou linhas) que não existem em todos os DataFrames que estão sendo unidos.
*   **`join='outer'` (Padrão):** É a "união". O resultado terá todas as colunas de todos os DataFrames. Se um DataFrame não possuir uma coluna que o outro tem, o Pandas preencherá esses espaços vazios com `NaN` (valores nulos).
*   **`join='inner'`:** É a "interseção". O resultado conterá apenas as colunas que aparecem em **todos** os DataFrames selecionados. Colunas que existem em apenas um deles são descartadas.

### 3. `axis` (O parâmetro mais relevante)
Mencionado nos títulos das fontes como a forma de escolher entre unir linhas ou colunas:
*   **`axis=0` (Padrão):** Une os dados verticalmente (adiciona linhas).
*   **`axis=1`:** Une os dados horizontalmente (adiciona colunas, colocando um DataFrame ao lado do outro).

### 4. `set_index` (Método relacionado)
Embora você tenha mencionado como uma função do `concat`, o **`set_index`** é, na verdade, um método separado do Pandas usado para definir uma coluna existente como o novo índice do DataFrame. Ele é frequentemente usado **antes** de operações de união (como o `join` ou `merge`) para garantir que os dados se alinhem corretamente por uma chave específica (como um ID de cliente).

### Outro parâmetro relevante: `keys`
Um parâmetro adicional muito útil é o **`keys`**. Ele permite criar um **Índice Hierárquico** (MultiIndex). Ao passar uma lista de nomes para `keys`, você consegue identificar no DataFrame final de qual DataFrame original cada linha veio, criando uma espécie de "etiqueta" para os blocos de dados unidos.

## Join

### 1. União Básica de Linhas (Vertical)
O padrão do `concat` é empilhar os dados um abaixo do outro (`axis=0`).
*   **Exemplo:** `pd.concat([df1, df2])`
*   **O que acontece:** O Pandas une as linhas. Se houver colunas que existem em um DataFrame mas não no outro, ele preencherá os valores com `NaN`. Os índices originais são mantidos, o que pode causar repetições (como os índices 4 e 5 que apareciam em ambos os DFs no vídeo).

### 2. Uso do `ignore_index`
Para evitar índices repetidos e criar uma nova sequência numérica:
*   **Exemplo:** `pd.concat([df1, df2], ignore_index=True)`
*   **Resultado:** O índice final será uma sequência limpa de 0 até o total de linhas menos um.

### 3. Diferença entre `join='outer'` e `join='inner'`
*   **`join='outer'` (Padrão):** Retorna todas as colunas de ambos os DataFrames.
*   **`join='inner'`:** Retorna apenas as colunas que são **comuns** aos dois DataFrames (no exemplo do vídeo, apenas a 'coluna1').

### 4. União de Colunas (Horizontal)
Para colocar um DataFrame ao lado do outro, utiliza-se o eixo das colunas.
*   **Exemplo:** `pd.concat([df1, df2], axis=1)`
*   **Observação importante:** O Pandas alinha as linhas pelo **índice**. No vídeo, onde os índices eram iguais (4 e 5), os dados foram unidos na mesma linha; onde eram diferentes, gerou-se novos espaços com `NaN`.

### 5. Alinhamento com `set_index`
Se você quiser unir colunas lateralmente usando uma coluna específica como chave (em vez do índice numérico padrão):
*   **Exemplo:** `pd.concat([df1.set_index('coluna1'), df2.set_index('coluna1')], axis=1)`
*   **Resultado:** Os dados serão alinhados corretamente com base nos valores da 'coluna1'.

### 6. Verificação de Integridade (`verify_integrity`)
Serve para checar se existem índices duplicados que podem causar erros na análise.
*   **Exemplo:** `pd.concat([df1, df2], verify_integrity=True)`
*   **Resultado:** O Pandas retornará um erro se encontrar índices sobrepostos (como o 4 e 5 mencionados).

### 7. Identificação de Origem com `keys`
Cria um índice hierárquico (MultiIndex) para identificar de qual DataFrame veio cada informação.
*   **Exemplo:** `pd.concat([df1, df2], keys=['DF1', 'DF2'])`
*   **Resultado:** O DataFrame terá uma "etiqueta" lateral indicando se os dados vieram do 'DF1' ou 'DF2'. Você também pode nomear esses níveis de índice (ex: `names=['Origem', 'ID']`).

## Merge

### 1. Funcionamento Padrão e o Parâmetro `how`
Diferente de outras funções, o `merge` identifica automaticamente colunas com nomes iguais para realizar a união. O parâmetro **`how`** define qual critério de união será usado:
*   **`inner` (padrão):** Retorna apenas os elementos que existem em ambos os DataFrames.
*   **`left`:** Usa o DataFrame da esquerda como base, mantendo todos os seus registros.
*   **`right`:** Usa o DataFrame da direita como base.
*   **`outer`:** Realiza uma união completa, trazendo todos os dados de ambos os DataFrames.

### 2. Especificando Colunas ou Índices
*   **`on`**: Usado para explicitar qual coluna será a chave da união (os nomes devem ser iguais em ambos os DFs).
*   **`left_on` e `right_on`**: Essencial quando você quer unir por colunas que possuem **nomes diferentes** nos dois DataFrames.
*   **`left_index` e `right_index`**: Quando definidos como `True`, permitem que a união seja feita utilizando o **índice** em vez de colunas.

### 3. Tratamento de Colunas Duplicadas (`suffixes`)
Se os DataFrames possuírem colunas com o mesmo nome que não fazem parte da união, o Pandas adiciona sufixos automaticamente (como `_x` e `_y`).
*   O parâmetro **`suffixes`** permite personalizar esses nomes, passando uma lista, por exemplo: `suffixes=['_df1', '_df2']`.

### 4. Parâmetro `indicator`
Ao definir **`indicator=True`**, o Pandas cria uma coluna extra (geralmente chamada `_merge`) que indica a origem do dado: se ele estava presente em ambos os DataFrames, apenas no da esquerda ou apenas no da direita.

### 5. Simulação de "PROCV" e Validação
O autor demonstra como usar o `merge` para realizar uma operação semelhante ao **PROCV (VLOOKUP)** do Excel, unindo uma tabela de referência a uma tabela com dados repetidos usando `how='left'`.
*   **`validate`**: Este parâmetro é usado para garantir a integridade da operação. Ele verifica se a relação é de um-para-um (`1:1`), um-para-muitos (`1:m`), muitos-para-um (`m:1`) ou muitos-para-muitos (`m:m`). Se a regra for violada (por exemplo, pedir `1:1` em dados duplicados), o Pandas retornará um erro.

## Append

### 1. Funcionamento Básico
O método `append` é utilizado especificamente para **adicionar ou unir linhas**. Segundo o autor, ele funciona de forma idêntica ao `pd.concat` quando este é usado para empilhar dados verticalmente.
*   **Exemplo:** `df1.append(df2)`
*   **O que acontece:** O `df2` é encaixado logo abaixo do `df1`. Se houver colunas que existem em apenas um dos DataFrames, o Pandas criará essas colunas no resultado final e preencherá os valores ausentes com `NaN`.

### 2. Tratamento de Índices (`ignore_index`)
Assim como no `concat`, o `append` preserva os índices originais por padrão, o que pode resultar em índices repetidos no novo DataFrame.
*   **Solução:** Use o parâmetro **`ignore_index=True`**.
*   **Resultado:** O Pandas descarta os índices antigos e cria uma nova sequência numérica de 0 até o total de linhas do novo DataFrame.

### 3. Persistência dos Dados (Não é *In-Place*)
É importante notar que o método `append` **não altera o DataFrame original** (não é *in-place*).
*   Para salvar as alterações, você deve atribuir o resultado a uma variável:
    *   **Exemplo:** `df1 = df1.append(df2)`.

### 4. Adicionando uma Única Linha com Dicionário
Além de unir dois DataFrames, você pode usar o `append` para inserir uma **única linha nova** utilizando um dicionário.
*   **Como fazer:** A chave do dicionário deve ser o nome da coluna e o valor deve ser o dado que você deseja inserir.
*   **Exemplo:** `df1.append({'coluna1': 'valor_h'}, ignore_index=True)`.
*   **Regra obrigatória:** Ao usar um dicionário, é **obrigatório** passar o parâmetro `ignore_index=True`, caso contrário, o Pandas retornará um erro por não saber qual índice atribuir à nova linha.

Para otimizar o uso do append, é recomendado unir apenas as linhas do seu DF, para evitar possíveis erros.

# **Value Counts**

O **`value_counts()`** é uma função do Pandas utilizada para **contar a frequência de valores únicos** em uma coluna (Series) de um DataFrame. Ele permite saber, por exemplo, quantas vezes cada número ou categoria se repete em seus dados.

Essa função pode ser acompanhada dos seguintes parâmetros>

### 1. `df['col3'].value_counts()`
Este comando básico conta as ocorrências de cada item na **coluna 3**. Por padrão, ele retorna os valores e a quantidade de vezes que cada um aparece, mas ignora valores nulos.

### 2. Parâmetros da função
*   **`dropna=False`**: Por padrão, o Pandas exclui valores nulos da contagem (`dropna=True`). Ao definir como `False`, a função passará a **exibir e contar também os valores vazios (NaN)**.
*   **`normalize=True`**: Em vez de mostrar a contagem absoluta (1, 2, 3...), este parâmetro retorna a **frequência relativa ou percentual** (ex: 0.6 para 60%). É muito útil para saber a proporção de cada valor no conjunto total.
*   **`sort=True`**: Define se o resultado deve ser **ordenado** pela quantidade de repetições. O padrão é `True`, o que organiza os dados do valor que mais se repete para o que menos aparece.
*   **`ascending=False`**: Trabalha em conjunto com o `sort`. Quando é `False` (padrão), a ordem é **decrescente** (do maior para o menor). Se alterado para `True`, a lista mostrará primeiro os itens com menos repetições.

### 3. Agrupamento de dados numéricos
*   **`df['col1'].value_counts(bins=4)`**: Quando usado em colunas numéricas, o parâmetro `bins` **agrupa os valores em intervalos iguais** (neste caso, 4 grupos) e conta quantos elementos existem dentro de cada faixa de valores.
*   **`pd.cut(df['col1'], 4).value_counts()`**: Este comando produz o **mesmo resultado** que o anterior. A função `pd.cut` divide os dados da 'col1' em 4 intervalos e o `value_counts()` realiza a contagem de itens em cada um desses intervalos gerados.

# Substituindo um valores no DF

### 1. Substituição em todo o DataFrame
Se você deseja buscar um valor em todas as colunas e substituí-lo, a estrutura básica é:
*   **`df.replace(valor_antigo, valor_novo)`**: Substitui todas as ocorrências de um valor por outro.
*   **Usando Dicionários**: `df.replace({valor_antigo: valor_novo})`. É útil para mapear várias substituições de uma vez.
*   **Usando Listas**: `df.replace([lista_antigos], [lista_novos])`. Substitui os itens da primeira lista pelos correspondentes na segunda.

### 2. Substituição em uma Coluna Específica
Para restringir a alteração a apenas uma parte dos dados:
*   **`df['nome_coluna'].replace(antigo, novo)`**: Altera apenas os valores dentro da coluna selecionada.

### 3. Substituição de Partes de uma String (Texto)
Se você precisar substituir apenas um caractere ou uma parte de uma palavra, o `replace` comum pode não funcionar a menos que você especifique que está lidando com padrões de texto:
*   **Com Regex**: `df['coluna'].replace('parte_antiga', 'parte_nova', regex=True)`. O parâmetro **`regex=True`** permite que o Pandas identifique pedaços de texto dentro das células.
*   **Com o acessor `.str`**: `df['coluna'].str.replace('parte_antiga', 'parte_nova')`. Esta é uma forma direta de tratar a coluna como string para substituições parciais.
    *   *Dica:* Ao usar `.str.replace()`, você pode usar o parâmetro **`n=1`** para limitar a substituição a apenas a primeira ocorrência encontrada em cada linha.

### 4. Substituição por Valores Vizinhos (Métodos)
Você também pode substituir valores baseando-se nos dados ao redor:
*   **`method='ffill'`**: Preenche com o **próximo valor** disponível.
*   **`method='bfill'`**: Preenche com o **valor anterior**.
*   **`limit`**: Você pode usar este parâmetro para restringir quantas vezes essa substituição por vizinhança deve ocorrer.

### 5. Tornando as Alterações Permanentes
É importante notar que, por padrão, o Pandas não salva a alteração no DataFrame original, apenas exibe o resultado. Para que a mudança seja gravada, você deve usar:
*   **`inplace=True`**: Adicione este parâmetro dentro da função (ex: `df.replace(5, 10, inplace=True)`) para modificar o objeto diretamente.

### 1. Utilizando a função `where()`
A função `where()` funciona de uma forma que pode parecer contra-intuitiva no início: ela **mantém** os valores onde a condição é verdadeira e **substitui** onde a condição é **falsa**.
*   **Estrutura:** `df.where(condicao, valor_substituto)`
*   **Exemplo:** Se você quiser que todos os valores que **não** são maiores que 10 sejam substituídos pela palavra "Oi", o código seria `df.where(df > 10, 'Oi')`.

### 2. Utilizando a função `mask()`
A função `mask()` é o exato oposto da `where()`. Ela **substitui** os valores onde a condição é **verdadeira**.
*   **Estrutura:** `df.mask(condicao, valor_substituto)`
*   **Exemplo:** Para substituir todos os valores maiores que 10 por "Oi", você usaria `df.mask(df > 10, 'Oi')`.

### 3. Substituição por valores dinâmicos (como o máximo da linha)
Você também pode substituir os dados por outros valores do próprio DataFrame, como o maior valor de uma linha ou coluna.
*   Para substituir valores que não atendem a uma condição pelo maior valor daquela linha, usa-se: `df.where(df > 10, df.max(axis=1), axis=0)`.
*   Nesse caso, o parâmetro **`axis=1`** dentro do `max()` indica que a busca é por linha, enquanto o **`axis=0`** no `where()` indica que a operação deve ser aplicada em cada coluna.

### 4. Outras alternativas
Embora o vídeo foque em `where()` e `mask()` por serem mais diretas, as fontes mencionam que o método **`.loc`** é uma forma extremamente versátil e poderosa de realizar substituições condicionais, permitindo resultados que as outras funções podem não alcançar.

**Dica importante:** Ao usar `where()` ou `mask()`, preste atenção se as dimensões dos dados batem (especialmente ao usar funções como `max()`), caso contrário o resultado pode ficar desconfigurado.
